### preprocess_emergencias
Sirve particularmente para los `.csv` de la parte de emergencias, ya que los de ambulatorio tienen otra estructura. Para utilizarla, ejecutar lo siguiente:
  * `from hmn_functions import preprocess_emergencias`, y luego
  * `dataframe` = preprocess_emergencias('directorio/donde/se/encuentra/el/.csv')

> *Nota:* sustituir la palabra '*dataframe*' por el nombre que se le quiera dar a la variable (por ejemplo, emergencias_enero [si fuera de las atenciones por emergencias en enero], o emergencias_2021 [si fueran todas las atenciones ingresadas por emergencias en el año 2021].


```python
csv_file = "emergency_masked.csv"
```

```python
# Ejemplo:
from hmn_functions import preprocess_emergencias
emergencia = preprocess_emergencias(csv_file)
emergencia.head(10)
```

### top_20_professionals
Imprime una tabla y un gráfico de barras de los primeros 20 profesionales con más atenciones de todas las secciones en el período del dataframe ingresado. Y luego una tabla y un gráfico de barra de los 20 profesionales con más atenciones de cada sección en el período del dataframe ingresado. Para utilizarla, ejecutar lo siguiente:
  * `from hmn_functions import top_20_professionals`, y luego
  * top_20_professionals(dataframe)

Por defecto imprime el gráfico de barras **total**. Para elegir criterios cuando se llama la función agregar las siguientes variables:
* `por_seccion = True` --> para incluir gráficos divididos por sección. De lo contrario, `por_seccion = False`


```python
# Ejemplo:
from hmn_functions import top_20_professionals
top_20_professionals(emergencia, por_seccion=False)
```

### atenciones
Imprime una tabla y/o un gráfico de torta de las atenciones del dataframe en el período del dataframe ingresado, divididas por servicio o por sección. Para utilizarla, ejecutar lo siguiente:
  * `from hmn_functions import atenciones_por_seccion`, y luego
  * atenciones(dataframe)

Por defecto imprime el gráfico de barras por servicio. Para elegir criterios cuando se llama la función agregar las siguientes variables:
* `por_seccion = True` (por defecto) --> para incluir gráficos divididos por sección. De lo contrario, `por_seccion = False`
* `por_servicio = True` --> para incluir gráficos divididos por servicio. De lo contrario, `por_servicio = False`
* `barra = True` (por defecto)  --> para incluir gráfico de barra. De lo contrario, `barra = False`
* `torta = True` --> para incluir gráfico de torta. De lo contrario, `torta = False`


```python
# Ejemplo:
from hmn_functions import atenciones
atenciones(emergencia, por_seccion=True, por_servicio=False, torta=False, barra=True)
```
